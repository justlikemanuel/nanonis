```python
import socket

HOST = '127.0.0.1'
PORT = 6501  # Dein Nanonis TCP-Port

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    try:
        s.connect((HOST, PORT))
        print(f"Verbunden mit {HOST}:{PORT}")
        


    except ConnectionRefusedError:
        print(f"Verbindung zu {HOST}:{PORT} abgelehnt")
    except Exception as e:
        print("Fehler:", e)
```
